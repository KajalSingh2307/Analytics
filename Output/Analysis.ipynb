### Step 13: Handled Outliers in Numerical Columns Using the IQR Method

- In this step, outliers in key numerical columns were addressed to improve data quality and ensure robust analysis.
- The Interquartile Range (IQR) method was used to detect and remove outliers. The IQR is calculated as the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of a dataset.
- For each numerical column ('price', 'mileage', 'tax', 'engineSize', 'Quantity', 'Pct'), the following process was applied:
    - Compute Q1 (25th percentile) and Q3 (75th percentile).
    - Calculate the IQR as Q3 - Q1.
    - Determine the lower bound as Q1 - 1.5 * IQR and the upper bound as Q3 + 1.5 * IQR.
    - Filter out rows where the column value falls outside these bounds, as these are considered outliers.
- This method helps to remove extreme values that could skew statistical analyses and visualizations, leading to more accurate and reliable insights.
- After outlier removal, the DataFrame was reset to ensure a continuous index.
- Handling outliers is crucial for many machine learning algorithms and statistical models, as it prevents undue influence from anomalous data points.

```python
# Handle outliers using the IQR method for numerical columns
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

for col in ['price', 'mileage', 'tax', 'engineSize', 'Quantity', 'Pct']:
    df = remove_outliers(df, col)
df.describe()
```

### Step 15: Analyzed Average Selling Price by Brand and Visualized It

- In this step, the analysis focused on understanding how the average price varies across different car brands (manufacturers).
- The DataFrame (df) was grouped by the 'Make' column, which represents the car brand. For each brand, the mean of the 'price' column was calculated using the groupby and mean functions.
- The resulting Series, containing the average price for each brand, was sorted in descending order to highlight those with the highest average prices at the top.
- To facilitate a clear comparison, a bar plot was created using the pandas plot method. The plot displays car brands on the x-axis and their corresponding average prices on the y-axis.
- This visualization quickly identifies premium brands as well as more affordable options.
- The analysis provides insights into brand value and market positioning, benefiting both buyers and sellers.


```python
avg_price_by_make = df.groupby('Make')['price'].mean().sort_values(ascending=False)
avg_price_by_make.plot(kind='bar', figsize=(12, 6))
plt.ylabel('Average Price')
plt.title('Average Car Price by Make')
plt.show()
```
